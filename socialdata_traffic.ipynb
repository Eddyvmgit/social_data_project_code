```python
###-------------------------------------------------------------------------------------------------------------------
#          Loading File and Viewing Data 
###-------------------------------------------------------------------------------------------------------------------

collision_df = pd.read_csv("Motor_Vehicle_Collisions_-_Crashes.csv")
collision_df.count()
```

```python
###-------------------------------------------------------------------------------------------------------------------
#          Cleaning Data 
###-------------------------------------------------------------------------------------------------------------------

# tranform to datetime
collision_df['CRASH DATE'] = pd.to_datetime(collision_df['CRASH DATE'], format='%m/%d/%Y')

# check for null values
collision_df = collision_df[~pd.isnull((collision_df['LATITUDE'])) & (~pd.isnull(collision_df['LONGITUDE']))]
collision_df = collision_df[(round(collision_df['LATITUDE']) != 0) & (round(collision_df['LONGITUDE']) != 0)]
collision_df = collision_df[~pd.isnull((collision_df['CRASH DATE'])) & (~pd.isnull(collision_df['CRASH TIME']))]
collision_df = collision_df[~pd.isnull((collision_df['NUMBER OF PERSONS KILLED']))]
collision_df = collision_df[~pd.isnull((collision_df['CONTRIBUTING FACTOR VEHICLE 1']))]

# check for incomplete years
print(collision_df['CRASH DATE'].min())
collision_df = collision_df[(collision_df['CRASH DATE'].dt.year != 2012) & (collision_df['CRASH DATE'].dt.year != 2023)]
```

```python
###-------------------------------------------------------------------------------------------------------------------
#          Summary Statistics
###-------------------------------------------------------------------------------------------------------------------

print(len(collision_df))
```

```python
###-------------------------------------------------------------------------------------------------------------------
#          Cluster all Accident Records to group them around Intersections
###-------------------------------------------------------------------------------------------------------------------

# https://datascience.stackexchange.com/a/848 - DBSCAN/OPTICS proposed algorith for LAT/LON Clustering
# https://www.youtube.com/watch?v=QsGOoWdqaT8&ab_channel=IGOT-ULisboa%2FCEG
# Rounding was necessary for computation time (might include slight inacccuracies)

from sklearn.cluster import DBSCAN
array_for_clustering = [[round(xy[0], 3), round(xy[1], 3)] for xy in zip(collision_df["LATITUDE"], collision_df["LONGITUDE"])]

# run algorithm with small eps distance threshold to get intersection level range
clustering = DBSCAN(min_samples=2, eps = 0.0001).fit(array_for_clustering)
```

```python
# check size of non clustered elements
print(len(clustering.labels_[clustering.labels_ == -1]))

# check number of clusters
print(np.unique(clustering.labels_))

# check size of clusters
counts, bins = np.histogram(clustering.labels_, bins=len(clustering.labels_), range=None, normed=None, weights=None, density=None)
plt.stairs(counts, bins)
plt.xlabel("Cluster Group")
plt.ylabel("Size")
```

```python
# add clustering labels
collision_df["Cluster"] = clustering.labels_
```

```python
# Check if there are relevant non-clustered elements (e.g high injury rate)
non_clustered_elements = collision_df[collision_df["Cluster"] == -1]
non_clustered_elements = non_clustered_elements.sort_values(by=['NUMBER OF PERSONS INJURED'], ascending=False)
non_clustered_elements = non_clustered_elements["NUMBER OF PERSONS INJURED"]
print(non_clustered_elements)
```

```python
collision_df_clustered = collision_df[collision_df["Cluster"] != -1]
```

```python
# group clusters to find most dangerous ones based on accident numbers
collision_df_clustered_grouped = collision_df_clustered.groupby('Cluster').agg({'COLLISION_ID':'count', 'NUMBER OF PERSONS KILLED':'sum', 'NUMBER OF PERSONS INJURED':'sum'})
collision_df_clustered_grouped = collision_df_clustered_grouped.sort_values(by=['COLLISION_ID'], ascending=False)

# get the 50 clusters with the highest accident count
ten_most_dangerous_clusters = collision_df_clustered_grouped.iloc[0:10]
ten_most_dangerous_clusters.columns = ["# Accidents", "# NUMBER OF PERSONS KILLED", "NUMBER OF PERSONS INJURED"]


clusters_of_interest = ten_most_dangerous_clusters.index

from tabulate import tabulate
print(tabulate(ten_most_dangerous_clusters, tablefmt="pipe", headers="keys"))
```

```python
# reduce original df to only these clustered accidents
final_collision_df_clustered = collision_df_clustered[collision_df_clustered["Cluster"].isin(clusters_of_interest)]
```

```python
###-------------------------------------------------------------------------------------------------------------------
#          Creating Interactive Bokeh plot for the 10 most dangerous clusters
###-------------------------------------------------------------------------------------------------------------------

import pandas as pd
from bokeh.palettes import TolRainbow
from bokeh.plotting import figure, show
from datetime import timedelta
from bokeh.io import output_file, show,save
from bokeh.layouts import gridplot
from bokeh.models import ColumnDataSource, Legend

# change long label for legend
final_collision_df_clustered["CONTRIBUTING FACTOR VEHICLE 1"] = final_collision_df_clustered["CONTRIBUTING FACTOR VEHICLE 1"].replace("Pedestrian/Bicyclist/Other Pedestrian Error/Confusion", "Pedestrian Error/Confusion")

# upper and lower boarder for plot ranges
upper_boarder = pd.to_datetime("06/01/2022", format='%m/%d/%Y')
lower_boarder = pd.to_datetime("06/01/2012", format='%m/%d/%Y')

# map each relevant category to a strict color to have color unity for subplots (only these categories are relevant in the top 50 clusters)
all_relevant_categories_included = ['Turning Improperly', 'Fatigued/Drowsy', 'View Obstructed/Limited', 'Obstruction/Debris', 'Passenger Distraction', 'Pedestrian Error/Confusion', 'Other Vehicular', 'Passing or Lane Usage Improper', 'Reaction to Uninvolved Vehicle', 'Pavement Slippery', 'Physical Disability', 'Unsafe Lane Changing', 'Following Too Closely', 'Driver Inattention/Distraction', 'Failure to Yield Right-of-Way', 'Traffic Control Disregarded', 'Unspecified', 'Unsafe Speed', 'Passing Too Closely', 'Driver Inexperience', 'Oversized Vehicle']
ensure_color_unity = dict(zip(all_relevant_categories_included, TolRainbow[21]))

plot_array = []
check_relevant_categories = []

# create a plot for the 10 clusters with the most injuries
for cluster in clusters_of_interest[0:10]:
    
    # get all accidents for current cluster
    current_cluster_df = final_collision_df_clustered[final_collision_df_clustered['Cluster'] == cluster]
    current_cluster_df['CRASH DATE'] = pd.to_datetime(current_cluster_df['CRASH DATE'], format='%m/%d/%Y')
    
    # create df for the barplot of the current cluster
    df_bar_plot = current_cluster_df.groupby([current_cluster_df['CRASH DATE'].dt.year]).agg(
        {'COLLISION_ID': 'count',
            'NUMBER OF PEDESTRIANS INJURED':'sum', 'NUMBER OF CYCLIST INJURED':'sum',
            'NUMBER OF MOTORIST INJURED':'sum'
        }).reset_index()

    # group pedestrians and cyclists
    df_bar_plot["NUMBER OF WALKER/CYCLISTS INJURED"] = df_bar_plot["NUMBER OF PEDESTRIANS INJURED"] +  df_bar_plot["NUMBER OF CYCLIST INJURED"]

    # reduce df to relevant columns
    df_bar_plot = df_bar_plot[['CRASH DATE', 'COLLISION_ID', 'NUMBER OF MOTORIST INJURED', 'NUMBER OF WALKER/CYCLISTS INJURED']]

    # update column causing_factors
    df_bar_plot.columns = ["CRASH DATE", "# ACCIDENTS", "# MOTORIST INJURED", "# WALKER/CYCLISTS INJURED"]

    # create own colum for year (else the hover value will be bugged using normal datetime)
    df_bar_plot['YEAR'] = df_bar_plot['CRASH DATE']

    # format crash date to datetime
    df_bar_plot['CRASH DATE'] = pd.to_datetime(df_bar_plot['CRASH DATE'], format='%Y')

    # define hover text
    tooltips = [
        ("YEAR", "@{YEAR}"),
        ("# ACCIDENTS", "@{# ACCIDENTS}"),
        ("# MOTORIST INJURED", "@{# MOTORIST INJURED}"),
        ("# WALKER/CYCLISTS INJURED", "@{# WALKER/CYCLISTS INJURED}")
    ]

    # bar plot
    b = figure(width=850, height=300, x_axis_type="datetime", tooltips = tooltips,  x_range=(lower_boarder, upper_boarder))
    b.xaxis.axis_label = 'Year'
    b.yaxis.axis_label = 'Incidents'
    b.title.text = 'Yearly Number of Accidents and Injuries - Cluster: ' + str(cluster)
    b.title.text_font_size = '12pt'

    # predefine categories for stacked-bar plot and their relevant colors
    categories = ["# WALKER/CYCLISTS INJURED", "# MOTORIST INJURED","# ACCIDENTS"]
    colors = ["#e84d60", "#718dbf", "#c9d9d3"]

    # plot bar
    v = b.vbar_stack(categories, x="CRASH DATE", width=timedelta(days=120), color=colors, source=ColumnDataSource(df_bar_plot))
    

    # update bar legend
    legend = Legend(items=[(x, [v[i]]) for i, x in enumerate(categories)], location=(-80, 170))
    b.add_layout(legend, 'right')
    b.legend.label_text_font_size = '8pt'
    b.legend.label_text_font_style = "bold"
    
    # create different df for line plot
    df_line_plot = current_cluster_df.groupby([current_cluster_df['CONTRIBUTING FACTOR VEHICLE 1']]).agg({'NUMBER OF PERSONS INJURED':'sum'}).reset_index()
    df_line_plot = df_line_plot.sort_values(by=['NUMBER OF PERSONS INJURED'], ascending=False)

    # filter for 10th most relevant accident causes for that cluster
    df_line_plot = df_line_plot.iloc[0:10]

    # hover labels for line plot
    tooltips_2 = [
        ("YEAR", "@{YEAR}"),
        ("CAUSE", "@{CONTRIBUTING FACTOR VEHICLE 1}"),
        ("# Accidents", "@{COLLISION_ID}")
    ]

    # line plot
    p = figure(width = 1000, height = 300, x_axis_type = "datetime", x_range = (lower_boarder, upper_boarder), tooltips = tooltips_2)
    p.title.text = 'Yearly Number of Accidents for each Causing Factor - Cluster: ' + str(cluster)
    p.title.text_font_size = '12pt'
    p.xaxis.axis_label = 'Year'
    p.yaxis.axis_label = 'Accidents'

    items = []
    for causing_factor in np.array(df_line_plot):

        # not relevant
        check_relevant_categories.append(causing_factor[0])

        # extract mapped color
        color = ensure_color_unity[causing_factor[0]]

        # prepare line for a specific factor of a specific cluster
        df_line_plot_for_specific_factor = current_cluster_df[current_cluster_df["CONTRIBUTING FACTOR VEHICLE 1"] == causing_factor[0]]
        df = df_line_plot_for_specific_factor.groupby(df_line_plot_for_specific_factor["CRASH DATE"].dt.year).agg({
             "COLLISION_ID":"count", 
             "CONTRIBUTING FACTOR VEHICLE 1": lambda x: str(set(x))[2:-2]
        }).reset_index()

        # year again for hover text (else bugged with datetime)
        df['YEAR'] = df['CRASH DATE']

        #transform date to datetime
        df['CRASH DATE'] = pd.to_datetime(df['CRASH DATE'], format='%Y')
        
        #plot line and safe reverence
        t = p.line(y = 'COLLISION_ID', x = 'CRASH DATE', source=ColumnDataSource(df), line_width = 2, color = color, alpha = 0.8)
        
        
        # for legend
        items.append((causing_factor[0], [t]))

    # misc and interactive legend
    legend = Legend(items = items, location = (0, 3)) 
    p.add_layout(legend, 'right') 
    p.legend.click_policy="hide"
    p.legend.label_text_font_size = '10pt'
    p.legend.label_text_font_style = "bold"

    # add plots to array
    plot_array.append(p)
    plot_array.append(b)

# group plots in cols
plot_col = gridplot(plot_array, ncols=2)

# show the results
show(plot_col)
output_file(filename="bokeh_plot.html", title="Static HTML file")
print(set(check_relevant_categories))
```

```python
collision_df = pd.read_csv("Motor_Vehicle_Collisions_-_Crashes.csv")

# Select the variables of interest
vars_of_interest = ["CRASH DATE", "CRASH TIME", "NUMBER OF PERSONS INJURED", "NUMBER OF PERSONS KILLED",
                    "NUMBER OF PEDESTRIANS INJURED", "NUMBER OF PEDESTRIANS KILLED", 
                    "NUMBER OF CYCLIST INJURED", "NUMBER OF CYCLIST KILLED",
                    "NUMBER OF MOTORIST INJURED", "NUMBER OF MOTORIST KILLED"]
collision_df = collision_df[vars_of_interest]

# Print the first few rows of the resulting dataframe
collision_df.head()

# Convert "CRASH DATE" to datetime format
collision_df["CRASH DATE"] = pd.to_datetime(collision_df["CRASH DATE"], format="%m/%d/%Y")

# Aggregate the data by day
df_daily = collision_df.groupby('CRASH DATE').sum()

# Adding a total number of incidents
df_daily['TOTAL incidents'] = df_daily.sum(axis=1)


df_daily
```

```python
df_daily_compressed = df_daily[["TOTAL incidents"]].reset_index()
df_daily_compressed.rename(columns={"TOTAL incidents": "Total Incidents", "CRASH DATE": "Date"}, inplace=True)
df_daily_compressed = df_daily_compressed.loc[(df_daily_compressed["Date"] >= "2013-01-01") & (df_daily_compressed["Date"] <= "2022-12-31")]
df_daily_compressed
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams.update({'font.size': 16})

# Group the data by month and year
df_monthly = df_daily_compressed.groupby([df_daily_compressed['Date'].dt.year, df_daily_compressed['Date'].dt.month])['Total Incidents'].sum()

# Convert the index to a list of tuples with (year, month) format
index = list(df_monthly.index.values)

# Extract the year and month from the index tuples
years = [i[0] for i in index]
months = [i[1] for i in index]

# Convert the months to radians
theta = [((m-1)/12)*2*np.pi + 1/24 for m in months]

# Create a polar plot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection': 'polar'})

# Set the angle ticks to represent the months
ax.set_xticks(np.arange(0, 2*np.pi, 2*np.pi/12))
ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Set the radial axis label
ax.set_rlabel_position(50)
ax.set_ylabel('Monthly Distribution of Accidents', labelpad=48, fontweight='bold')

# Define the colormap and normalize the year values
cmap = plt.cm.tab10
normalize = plt.Normalize(vmin=years[0], vmax=years[-1])

# Loop over each year and plot the data as a line with a different color
for year in sorted(set(years)):
    # Get the data for the current year
    year_data = df_monthly.loc[[years[i] == year for i in range(len(years))]]
    
    # Repeat the year_data.values 10 times (for each year) to match the length of theta
    year_data_values = np.tile(year_data.values, 10)
    
    # Normalize the year value to be in the range [0, 1]
    year_norm = normalize(year)
    
    # Get the color for the current year
    color = cmap(year_norm)
    
    # Plot the data as a line
    ax.plot(theta, year_data_values, color=color, label=str(year), alpha=0.5)

# Add a legend
ax.legend(title='Year', loc='upper left', bbox_to_anchor=(1.2, 1))

plt.show()

# Save the plot as a jpg picture
fig.savefig('total_incidents_per_month_circle.jpg')
```

```python
# Convert 'CRASH TIME' column to datetime type
collision_df['CRASH TIME'] = pd.to_datetime(collision_df['CRASH TIME'])

# Combine date and time information into a single datetime column
collision_df['CRASH TIME'] = pd.to_datetime(collision_df['CRASH DATE'].dt.date.astype(str) + ' ' + collision_df['CRASH TIME'].dt.time.astype(str))

# Aggregate the data by day
df_hourly = collision_df.groupby(pd.Grouper(key='CRASH TIME', freq='H')).sum()

# Adding a total number of incidents
df_hourly['TOTAL incidents'] = df_hourly.sum(axis=1)

df_hourly
```

```python
df_hourly_compressed = df_hourly.loc[:, ['TOTAL incidents']].reset_index()
df_hourly_compressed = df_hourly_compressed.loc[(df_hourly_compressed['CRASH TIME'].dt.year >= 2013) & (df_hourly_compressed['CRASH TIME'].dt.year <= 2022)]
df_hourly_compressed
```

```python
import matplotlib.pyplot as plt

# Calculate the average number of incidents per hour for each hour of the day
hourly_avg = df_hourly_compressed.groupby(df_hourly_compressed['CRASH TIME'].dt.hour)['TOTAL incidents'].mean()

# Plot the hourly average distribution
plt.plot(hourly_avg)
plt.xlabel('Hour of the day')
plt.ylabel('Average number of incidents')
plt.savefig('hourly_avg_incidents.jpg') # save the plot as a jpg image
plt.show()
```

```python
import seaborn as sns
plt.rcParams.update({'font.size': 14})

# Create a pivot table with rows representing the day of the week, columns representing the hour of the day, and values representing the average number of incidents
pivot_table = df_hourly_compressed.pivot_table(values='TOTAL incidents', index=df_hourly_compressed['CRASH TIME'].dt.hour, columns=df_hourly_compressed['CRASH TIME'].dt.dayofweek, aggfunc='mean')

# Set the order of days of the week
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Set the order of hours of the day
hour_order = [i for i in range(24)]

# Create the heatmap
fig, ax = plt.subplots(figsize=(6,5)) 
sns.heatmap(pivot_table, cmap='YlOrRd', xticklabels=day_order, yticklabels=hour_order)

# Set the title and axis labels
plt.title('Average number of incidents per hour and day')
plt.xlabel('Day of the week')
plt.ylabel('Hour')

# Show the plot
plt.show()



```
